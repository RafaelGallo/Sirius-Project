# Resultados da pesquisa do exoplaneta Kepler

**10.000 candidatos a exoplanetas examinados pelo Observatório Espacial Kepler**

**Contexto**

O Observatório Espacial Kepler é um satélite construído pela NASA que foi lançado em 2009. O telescópio é dedicado à busca de exoplanetas em sistemas estelares além do nosso, com o objetivo final de possivelmente encontrar outros planetas habitáveis ​​além do nosso. A missão original terminou em 2013 devido a falhas mecânicas, mas o telescópio está funcional desde 2014 em uma missão estendida "K2".

O Kepler verificou 1.284 novos exoplanetas em maio de 2016. Em outubro de 2017, havia mais de 3.000 exoplanetas confirmados no total (usando todos os métodos de detecção, incluindo os baseados em terra). O telescópio ainda está ativo e continua a coletar novos dados em sua missão estendida.

**Conteúdo**

Um registro cumulativo de todos os "objetos de interesse" observados pelo Kepler - basicamente, todos os cerca de 10.000 candidatos a exoplanetas sobre os quais Kepler fez observações.

**Dicionário de dados**

As colunas destacáveis de observação são:

**kepoi_name:** Um KOI é um alvo identificado pelo Projeto Kepler que exibe pelo menos uma sequência semelhante a trânsito dentro da fotometria de série temporal do Kepler que parece ser de origem astrofísica e inicialmente consistente com uma hipótese de trânsito planetário

**kepler_name:** Têm a intenção de indicar claramente uma classe de objetos que foram confirmados ou validados como planetas - um passo à frente da designação de candidato a planeta.

**koi_disposition:** A disposição na literatura em relação a este candidato a exoplaneta. Um de CANDIDATO, FALSO POSITIVO, NÃO DESCARTADO ou CONFIRMADO.

**koi_pdisposition:** A disposição que a análise de dados do Kepler tem em relação a este candidato a exoplaneta. Um de FALSO POSITIVO, NÃO DISPOSITIVO e CANDIDATO.

**koi_score:** Um valor entre 0 e 1 que indica a confiança na disposição KOI. Para CANDIDATOS, um valor mais alto indica mais confiança em sua disposição, enquanto para FALSOS POSITIVOS, um valor mais alto indica menos confiança nessa disposição.

# Sistema de coordenadas celestes

Esta é a 'latitude' e 'longitude' dos objetos, então este gráfico mostra sua posição no céu (é um pequeno pedaço no céu, não se engane.) O telescópio apontou para o mesmo pedaço do céu por toda a missão e a distribuição parecem como estão por causa disso.

A densidade das estrelas aumenta em um lado, que é o lado mais próximo do bojo da Via Láctea.

![Here's an image from the CCD array](https://archive.stsci.edu/files/live/sites/mast/files/home/missions-and-data/k2/_images/k2_explained_25nov_story.jpg?t=tn1200)

```python
confirmed = data[data.koi_disposition == 'CONFIRMED']
ra, dec = data.ra, data.dec
ra_c, dec_c = confirmed.ra, confirmed.dec

fig = plt.figure(figsize=(20, 15))
plt.scatter(ra, dec, s = 40, label = "Candidatos")
plt.scatter(ra_c, dec_c, s = 40, label = "Confirmado")
plt.title("Sistema de coordenadas celestes", fontsize = 30)
plt.xlabel("Planetas condidatos", fontsize = 30)
plt.ylabel("Planetas confirmados", fontsize = 30)
plt.legend()
```
